```python
file_path = "ucs-654-kaggle-hack-lab-exam-1-privateleaderboard-2025-01-22T07_48_17.csv"
df = pd.read_csv(file_path)
```

```python
def is_valid_team_name(team_name):
    # Split the TeamName and check for various formats
    parts = team_name.split("_")
    
    # Case 1: Check if the last part is a 9-digit roll number (firstname_lastname_rollno)
    if len(parts) >= 3 and parts[-1].isdigit() and len(parts[-1]) == 9:
        return True
    
    # Case 2: Check if the first part is a 9-digit roll number (rollno_firstname_lastname)
    elif len(parts) >= 3 and parts[0].isdigit() and len(parts[0]) == 9:
        return True
    
    # Case 3: Check if the second part is a 9-digit roll number (name_rollno)
    elif len(parts) == 2 and parts[1].isdigit() and len(parts[1]) == 9:
        return True
    
    # Case 4: Check if the first part is a 9-digit roll number (rollno_name)
    elif len(parts) == 2 and parts[0].isdigit() and len(parts[0]) == 9:
        return True
    
    return False  # Return False if none of the conditions are met

df_filtered = df[df['TeamName'].apply(is_valid_team_name)]
```

```python
# Define the weights and impacts
weights = {'Rank': 1, 'Score': 2, 'SubmissionCount': 3}
impacts = {'Rank': '-', 'Score': '+', 'SubmissionCount': '+'}
```

```python
# Normalize the data
def normalize_column(col, impact):
    if impact == '+':
        return (col - col.min()) / (col.max() - col.min())
    else:
        return (col.max() - col) / (col.max() - col.min())
```

```python
# Apply normalization
df_filtered.loc[:, 'NormalizedRank'] = normalize_column(df_filtered['Rank'], impacts['Rank'])
df_filtered.loc[:, 'NormalizedScore'] = normalize_column(df_filtered['Score'], impacts['Score'])
df_filtered.loc[:, 'NormalizedSubmissionCount'] = normalize_column(df_filtered['SubmissionCount'], impacts['SubmissionCount'])
```

```python
# Apply weights
df_filtered.loc[:, 'WeightedRank'] = df_filtered['NormalizedRank'] * weights['Rank']
df_filtered.loc[:, 'WeightedScore'] = df_filtered['NormalizedScore'] * weights['Score']
df_filtered.loc[:, 'WeightedSubmissionCount'] = df_filtered['NormalizedSubmissionCount'] * weights['SubmissionCount']
```

```python
df_filtered.head(15)
```

```python
# Ideal best and worst solutions
ideal_best = df_filtered[['WeightedRank', 'WeightedScore', 'WeightedSubmissionCount']].max()
ideal_worst = df_filtered[['WeightedRank', 'WeightedScore', 'WeightedSubmissionCount']].min()
```

```python
# Calculate Euclidean distances
df_filtered.loc[:, 'DistanceBest'] = np.sqrt(((df_filtered[['WeightedRank', 'WeightedScore', 'WeightedSubmissionCount']] - ideal_best) ** 2).sum(axis=1))
df_filtered.loc[:, 'DistanceWorst'] = np.sqrt(((df_filtered[['WeightedRank', 'WeightedScore', 'WeightedSubmissionCount']] - ideal_worst) ** 2).sum(axis=1))
```

```python
# Calculate TOPSIS score
df_filtered.loc[:, 'TopsisScore'] = df_filtered['DistanceWorst'] / (df_filtered['DistanceBest'] + df_filtered['DistanceWorst'])
```

```python
# Rank participants based on the TOPSIS score
df_filtered.loc[:, 'TopsisRank'] = df_filtered['TopsisScore'].rank(ascending=False)
```

```python
# Assign Marks
def assign_marks(rank, score, df_filtered):
    if rank <= len(df_filtered) * 0.15:
        marks = 2.0
    elif rank <= len(df_filtered) * 0.30:
        marks = 1.5
    elif rank <= len(df_filtered) * 0.45:
        marks = 1.0
    elif rank <= len(df_filtered) * 0.60:
        marks = 0.75
    elif rank <= len(df_filtered) * 0.75:
        marks = 0.5
    else:
        marks = 0.25

    # Check if the score is duplicated
    if df_filtered['Score'].tolist().count(score) > 1:
        marks = -10  # Deduct 10 marks if the score is duplicated

    return marks
```

```python
df_filtered['Marks'] = df_filtered.apply(lambda row: assign_marks(row['TopsisRank'], row['Score'], df_filtered), axis=1)
```

```python
# Update topsis_rank and topsis_score to null for students with Marks = -10
df_filtered.loc[df_filtered['Marks'] == -10, ['TopsisRank', 'TopsisScore']] = np.nan
```

```python
import re

def extract_roll_number(team_name):
    # Regular expression to capture the roll number from various formats
    match = re.search(r'(\d{1,9})', team_name)  # Assuming roll number is a numeric value of up to 4 digits
    return match.group(1) if match else None

df_filtered['Roll Number'] = df_filtered['TeamName'].apply(extract_roll_number)

df_filtered.loc[:, 'Name'] = df_filtered['TeamName']
```
